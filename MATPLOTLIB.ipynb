```python
from matplotlib import style

style.use("ggplot")

x=[5,8,10]
y=[12,16,6]
x2=[6,9,11]
y2=[6,15,7]

pit.plot(x,y,'g',label='line one',linewidth=5)
pit.plot(x2,y2,'c',label='line two',linewidth=5)

pit.title("info")
pit.ylabel("yaxis")
pit.xlabel("xaxis")

pit.legend()
pit.grid(True,color='R')

pit.show()
```

```python
from matplotlib import pyplot as plt
x=[2,5,8,9]
y=[3,8,4,7]
x1=[3,4,7,10]
y1=[4,3,8,5]

plt.bar(x,y,label="subjects",color='green')
plt.bar(x1,y1,label="completed ",color='blue')

plt.legend()

plt.title('info')
plt.xlabel(' subject code')
plt.ylabel('waitage of subject')

plt.show()

```

```python
from matplotlib import pyplot as plt

x=[2,4,5,7]
y=[4,3,6,1]
x1=[3,4,7,10]
y1=[4,3,8,5]


plt.scatter(x,y,label='scatter',color='red',marker='x')
plt.scatter(x1,y1,label='SSS',color='blue',marker='<',s=120)    #s is size of marker

plt.legend()

plt.show()
```

```python
from matplotlib import pyplot as plt

x=[20,30,88,45]
l=['english','maths','computer','physics']
col=['blue','y','r','g']
t=(0,0,0,0.03)

plt.title("PIE CHArT")
plt.pie( x, labels=l, colors=col, explode=t, autopct='%1.2f%%', shadow=True)      
#explode is used to highlight and it is written in tupel
#autopct is auto percentage  1.2f 2 is no of digits                                                                                                                  

plt.legend()

plt.show()
```

```python
from matplotlib import pyplot as plt

#subplot   is to display more plot or graph in a figure 

x=[1,2,3,4]
y=[2,2,2,2]

x1=[3,4,6,8]
y1=[3,4,5,6]

plt.subplot(2,2,1)              # subplot(nrows, ncols, index, **kwargs)                 
plt.plot(x,y,'r')


plt.subplot(2,2,3)
plt.plot(x,y1,'g')

plt.subplot(2,2,2)
plt.plot(x1,y1,'ro')

plt.subplot(2,2,4)
plt.bar(x1,y1,color='blue')

plt.show()

x=[1,2,3,4]
y=[2,2,2,2]

x1=[3,4,6,8]
y1=[3,4,5,6]

plt.subplot(4,1,1)                              
plt.plot(x,y,)


plt.subplot(4,1,2)
plt.plot(x,y1,'g')

plt.subplot(4,1,3)
plt.plot(x1,y1,'ro')

plt.subplot(4,1,4)
plt.bar(x1,y1,color='blue')

plt.show()

x=[1,2,3,4]
y=[2,2,2,2]

x1=[3,4,6,8]
y1=[3,4,5,6]

plt.subplot(3,2,1)                              
plt.plot(x,y,)


plt.subplot(3,2,2)
plt.plot(x,y1,'g')

plt.subplot(3,2,3)
plt.plot(x1,y1,'ro')

plt.subplot(3,2,6)
plt.bar(x1,y1,color='blue')

plt.show()
```
